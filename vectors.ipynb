```python
### Types of ways to compute Dot Product !!!

v1 = np.array([4, 5, 7, 9, 3])
v2 = np.array([0, -4, -3, 6, 6])

dp1 = sum(np.multiply(v1,v2))

dp2 = np.dot(v1 , v2)

dp3 = np.matmul(v1 , v2)

dp4 = 0

for i in range(0, len(v1)):
    dp4 = dp4 + v1[i] * v2[i]
    
print(dp1, dp2, dp3, dp4)
```

```python
# Distributive Property 
## a'(b + c ) = ( a + b )' * c ( In case of vectors this property holds true)

n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

res1 = np.dot(a , b+c)
res2 = np.dot(a, b ) + np.dot(a,c)

print(res1, res2)
```

```python
# associative property
## a'( b c ) = ( a b )' c  ( In case of vectors it is not associative)

n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

res1 = np.dot(a , np.dot(b,c))
res2 = np.dot(np.dot(a,b),c)

print([res1, res2])

### Special cases which holds true are where one of the vectors is 0 ; a == b == c 
```

```python
# length of a vector

v1 = np.array([1 , 2, 5, 3])

# method - 1
length = np.sqrt(np.dot(v1, v1))

print(length)

# method - 2
length = np.sqrt(np.sum(np.multiply(v1, v1)))

print(length)

# method - 3
length = np.linalg.norm(v1)

print(length)
```

```python
vector1 = np.array([2, 4, -3])
vector2 = np.array([0, -3, -3])

ang = np.arccos((np.dot(vector1, vector2)) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot([0, vector1[0]], [0, vector1[1]], [0, vector1[2]])
ax.plot([0, vector2[0]], [0, vector2[1]], [0, vector2[2]])

ax.set_xlim(-6,6)
ax.set_ylim(-6,6)
ax.set_zlim(-6,6)


plt.title(f'Angle between vectors : {ang:.2f} rand.')
plt.show()
```

```python
dp_algebra = np.dot(vector1, vector2)

dp_geometric = np.linalg.norm(vector1) * np.linalg.norm(vector2) * np.cos(ang)


print(dp_algebra, dp_geometric)
```

```python
# Cauchy-Schwarz Inequality

a = np.random.randn(5)

b = np.random.randn(5)

c = np.random.randn(1) * a


ab = np.abs(np.dot(a, b))

ac = np.abs(np.dot(a,c))


length_a = np.linalg.norm(a)

length_b = np.linalg.norm(b)

length_c = np.linalg.norm(c)


print(f'{ab:.4f}, {(length_a * length_b):.4f}')

print(f'Both values here are equal for a and c (Dependent vectors) : {ac:.4f}, {(length_a * length_c):.4f}')

```

### Dot Product between the Unit Vectors !!

#### take away is the Magnitude of the dot product of unit vectors is smaller than the length of the both vectors i.e ( Less than 1)

```python
import numpy as np

rand_v1 = np.array(np.random.randn(4))

rand_v2 = np.array(np.random.randn(4))

print(rand_v1, rand_v2)
```

```python
rand_v1_len = np.linalg.norm(rand_v1)

rand_v2_len = np.linalg.norm(rand_v2)

print(rand_v1_len, rand_v2_len)

rand_v1_len = np.sqrt(np.dot(rand_v1, rand_v1))  ## Another way to compute the length of the vector !

rand_v2_len = np.sqrt(np.dot(rand_v2, rand_v2))

print(rand_v1_len, rand_v2_len)
```

```python
rand_dp = np.dot(rand_v1, rand_v2)

print(np.abs(rand_dp))
```

```python
rand_norm_v1 = rand_v1 / rand_v1_len

rand_norm_v2 = rand_v2 / rand_v2_len

print("Unit vectors : ", rand_norm_v1, rand_norm_v2)

rand_norm_v1_len = np.linalg.norm(rand_norm_v1)

rand_norm_v2_len = np.linalg.norm(rand_norm_v2)

print("\nLength of the unit vectors : ", rand_norm_v1_len , rand_norm_v2_len)


rand_norm_dp = np.dot(rand_norm_v1, rand_norm_v2)

print("\nMagnitude of the unit vectors dot product : ", np.abs(rand_norm_dp))
```

### SPAN :

span ( {v<sub>1</sub>, v<sub>2</sub>.........v<sub>n</sub>}) = &alpha;<sub>1</sub>v<sub>1</sub> +  &alpha;<sub>2</sub>v<sub>2</sub> + ...... + &alpha;<sub>n</sub>v<sub>n</sub> , $\alpha \in \mathbb{R}$


Question : If you have a two vectors suppose <b>v</b> , <b>w</b>  and if you want to say that these vectors are in this span of <b>S</b> then we have to show that these vectors ( <b>v</b> , <b>w</b>  ) can be represent linearly with in the span vectors. 

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

S1 = np.array([1, 1, 0])
S2 = np.array([3, 2, 1])

v = np.array([1, 2, 0])
w = np.array([3, 2, 1])


s = np.linspace(-2, 2, 10)
t = np.linspace(-2, 2, 10)
S, T = np.meshgrid(s, t)

X = S * S1[0] + T * S2[0]
Y = S * S1[1] + T * S2[1]
Z = S * S1[2] + T * S2[2]


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.plot_surface(X, Y, Z, alpha=0.4, color='cyan')


ax.quiver(0, 0, 0, *S1, color='green')
ax.quiver(0, 0, 0, *S2, color='lime')


ax.quiver(0, 0, 0, *v, color='red')
ax.quiver(0, 0, 0, *w, color='blue')


ax.text(*S1, 'S1', color='green')
ax.text(*S2, 'S2', color='lime')
ax.text(*v, 'v', color='red')
ax.text(*w, 'w', color='blue')


ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Span of S1 and S2 (Plane) with vectors v and w")

plt.show()
```

#### Formal Defintiion of Linear Dependence : 
##### 0 = &lambda;<sub>1</sub>v<sub>1</sub> + &lambda;<sub>2</sub>v<sub>2</sub>  + ..........+&lambda;<sub>n</sub>v<sub>n</sub> ; Where $\lambda \in \mathbb{R}$ for atleast one &lambda; is not equal to 0 

#### Theorem : Maximum N independent Vectors in $\mathbb{R}$<sup>N</sup>

##### Any Set of M > N vectors in $\mathbb{R}$<sup>N</sup> is dependent 

##### Any Set of M <= N vectors in $\mathbb{R}$<sup>N</sup> could be Independent 

#### Steps to determine whether a set is Independent 

##### Step 1 : Count Vectors and Compare with $\mathbb{R}$<sup>N</sup>.
##### Step 2 : Check for 0<sup>'</sup>s in all the corresponding (or all) the elements.
##### Step 3 : Educated guess and Test
##### Step 4 : Matrix rank method
